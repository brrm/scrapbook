# OLS

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
X = df[['LSTAT', 'DIS', 'TAX', 'PTRATIO', 'RM']]
y = df['MEDV']
X = sm.add_constant(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, shuffle=True)
```

```python
import statsmodels.api as sm
```

```python
model = sm.OLS(exog=X_train, endog=y_train)
results = model.fit() #_regularized(L1_wt=1, alpha=0) # L1_wt=1 means Lasso
```

```python
results.summary()
```

```python
sns.scatterplot(x=y_train, y=results.resid)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(results.predict(X_test), y_test)
```

```python
fig = sm.graphics.plot_regress_exog(results, "LSTAT")
fig.tight_layout(pad=1.0)
```

```python
fig = sm.graphics.plot_fit(results, "LSTAT")
fig.tight_layout(pad=1.0)
```

```python
conf = results.get_prediction(X_test).summary_frame(alpha=0.05).join(X_test['LSTAT'])
conf.head(5)
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y_upper = y + 0.3
y_lower = y - 0.3

ax = sns.lineplot(x=x, y=y)
ax.fill_between(x, y_lower, y_upper, alpha=0.2, color='b', label='Confidence')
```
