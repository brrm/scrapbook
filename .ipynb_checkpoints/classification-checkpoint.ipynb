```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris(as_frame=True)['frame']
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
df['sepal area'] = df['sepal length (cm)'] * df['sepal width (cm)']
df['petal area'] = df['petal length (cm)'] * df['petal width (cm)']
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(8, 5))
axs = axs.flatten()

for i, col in enumerate(df.drop(['target'], axis=1).columns):
    sns.boxplot(data=df, y=col, ax=axs[i])
plt.tight_layout()
```

```python
sns.displot(df['sepal area'], kde=True, rug=True)
```

```python
sns.regplot(data=df, x='sepal length (cm)', y='sepal width (cm)')
```

```python
sns.scatterplot(data=df, x='sepal area', y='petal area', hue='target')
```

```python
from sklearn import decomposition
```

```python
X = df.drop(['target', 'petal area', 'sepal area'], axis=1)
y = df['target']
```

```python
pca = decomposition.PCA(n_components=3)
X_norm = (X - X.mean()) / X.std()
pca.fit(X_norm)
```

```python
sns.heatmap(pd.DataFrame(pca.components_, columns=X.columns), annot=True)
```

```python
sns.lineplot(data=pca.explained_variance_, marker='o')
```

```python
Z = pd.DataFrame(pca.transform(X_norm))
Z['target'] = df['target']
```

```python
sns.scatterplot(data=Z, x=0, y=1, hue='target')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn import svm
```

```python
svc = svm.SVC()
```

```python
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

```python
df_eval = X_test.copy()
df_eval['target'] = y_test
df_eval['pred'] = y_pred
df_eval['error'] = (df_eval['target'] != df_eval['pred'])
```

```python
sns.scatterplot(data=df_eval, x='petal length (cm)', y='sepal length (cm)', hue='target', style='error')
```
